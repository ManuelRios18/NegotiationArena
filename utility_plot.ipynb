```python
import os
import json
import matplotlib.pyplot as plt
import matplotlib.cm
```

```python
# exp_path = ".logs/1698311548790"
# exp_path = ".logs/1698310386029"
```

```python
def load_states(paths):
    states = {}
    for exp_path in paths:
        with open(os.path.join(exp_path,"state.json")) as f:
            states[exp_path] = json.load(f)
    return states

def get_in_trade_utility(state):
    agent_1_proposal_utility = [ eval(_['in_trade_utility'])[1] for _ in state[0] if 'in_trade_utility' in _]
    agent_2_proposal_utility = [ eval(_['in_trade_utility'])[2] for _ in state[1] if 'in_trade_utility' in _]
    return {1: agent_1_proposal_utility, 2: agent_2_proposal_utility }
```

```python
states = load_states([
        ".logs/1698311548790", ".logs/1698310386029"
])
```

```python
utilities = [get_in_trade_utility(_) for _ in states.values()]
```

```python
utilities
```

```python
color = matplotlib.cm.datad['tab10']['listed']
for util in utilities:
    for idx, _ in enumerate(util.values()):
        plt.plot(_, color=color[idx])
```
